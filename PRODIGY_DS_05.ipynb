```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("train.csv")

df['Age'].fillna(df['Age'].median(), inplace=True)

df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)

plt.figure()
df['Survived'].value_counts().plot(kind='bar')
plt.title("Survival Count")
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.ylabel("Number of Passengers")
plt.show()

plt.figure()
df.groupby('Sex')['Survived'].mean().plot(kind='bar')
plt.title("Survival Rate by Gender")
plt.xlabel("Gender")
plt.ylabel("Survival Rate")
plt.show()

plt.figure()
df.groupby('Pclass')['Survived'].mean().plot(kind='bar')
plt.title("Survival Rate by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Survival Rate")
plt.show()

plt.figure()
df['Age'].plot(kind='hist', bins=20)
plt.title("Age Distribution of Passengers")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()

plt.figure()
df.groupby('Embarked')['Survived'].mean().plot(kind='bar')
plt.title("Survival Rate by Embarkation Port")
plt.xlabel("Embarked Port")
plt.ylabel("Survival Rate")
plt.show()


print("\nKey Insights:")
print("Overall Survival Rate:", df['Survived'].mean())
print("Highest survival gender:", df.groupby('Sex')['Survived'].mean().idxmax())
print("Highest survival class:", df.groupby('Pclass')['Survived'].mean().idxmax())
print("Port with highest survival rate:", df.groupby('Embarked')['Survived'].mean().idxmax())
```
